```python
# Dependencies
import pandas as pd
from pathlib import Path
import re
```

```python
# Load in file
# Store filepath in a variable
movie_file = Path("Resources/DOHMH_Dog_Bite_Data.csv")
```

```python
# Read and display the CSV with Pandas
movie_file_df = pd.read_csv(movie_file)
movie_file_df.head(10)
```

```python
movie_file_df.replace(re.compile('.*pit bull.*', re.IGNORECASE), 'Pit Bull', inplace=True)
movie_file_df.replace(re.compile('.*pitbull.*', re.IGNORECASE), 'Pit Bull', inplace=True)
movie_file_df.head(10)
```

```python
movie_file_df["Age"].replace(re.compile('1/2'), '', inplace=True)

movie_file_df["Age"].replace(re.compile('\D'), '', inplace=True)

movie_file_df["New"] = movie_file_df["Species"].replace(re.compile('.*pitbull.*', re.IGNORECASE), 'Pit Bull', inplace=True)

#cols = ['Age', 'ZipCode']

#movie_file_df[cols] = movie_file_df[cols].apply(pd.to_numeric, errors='coerce')

movie_file_df['Age'] = pd.to_numeric(movie_file_df['Age'], errors='coerce')

movie_file_df.head(10)
```

```python
movie_file_df["DateOfBite"] = pd.to_datetime(movie_file_df["DateOfBite"])

movie_file_df["DateOfBite"] = movie_file_df["DateOfBite"].dt.strftime("%m-%d-%Y")

movie_file_df.head(10)
```

```python
movie_file_df.value_counts().head(10)
```

```python
# We only want IMDb data, so create a new table that takes the Film and all the columns relating to IMDB
imdb_df = movie_file_df[["FILM", "IMDB", "IMDB_norm",
                            "IMDB_norm_round", "IMDB_user_vote_count"]]
imdb_df.head()
```

```python
# We only like good movies, so find those that scored over 7, and ignore the norm rating
good_movies_df = movie_file_df.loc[movie_file_df["IMDB"] > 7, [
    "FILM", "IMDB", "IMDB_user_vote_count"]]
good_movies_df.head()
```

```python
# Find less popular movies--i.e., those with fewer than 20K votes
unknown_movies_df = good_movies_df.loc[good_movies_df["IMDB_user_vote_count"] < 20000, [
    "FILM", "IMDB", "IMDB_user_vote_count"]]
unknown_movies_df.head()
```

```python
# Finally, export this file to a spread so we can keep track of out new future watch list without the index
unknown_movies_df.to_excel("output/movieWatchlist.xlsx", index=False)
```
